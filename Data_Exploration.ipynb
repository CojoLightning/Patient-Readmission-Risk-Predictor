```python
#imports

#pandas is for data import
import pandas as pd

# math stuff
import numpy as np

# data vizualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
df = pd.read_csv("data/diabetic_data.csv")
with open("data/diabetic_data.csv", 'r') as f:
    mapping = f.read()
```

```python
#Initial Inspection of Data
print("\nInitial Data Shape:", df.shape)
# prints out as rows x columns
print("\nFirst 5 rows of the data:")
df.head(10)
```

```python
# replace missing data
df.replace('?', np.nan, inplace=True)
```

```python
# Calculate percentage of missing values per column
missing_info = df.isnull().sum() # counts missing values
missing_info = missing_info[missing_info > 0].sort_values(ascending=False) # filters to only columns with missing values and sorts
missing_percent = (missing_info / len(df)) * 100 # calculates percentage of missing values

missing_df = pd.DataFrame({
    'Missing Count': missing_info,
    'Missing %': missing_percent.round(2)
}) # creates a dataframe for better visualization

print("Columns with Missing Values ('?') or 'Unknown/Invalid':")
print(missing_df)
```
